```python
import pandas as pd
```

```python
df = pd.read_csv("spam_ham_dataset.csv")
df.head(3)
```

```python
df = df[['text','label_num']]
df.head(3)
```

```python
import nltk
nltk.download("stopwords")
```

```python
# Stopword kaldırma işlemi. -> the, at, on, is
# Stem işlemi (Kelimeleri köküne indirme) -> Running -> Run
# rUnning - RUNNING - ruNNing

import string
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords

# Python DERSLERİMİZ Hakkında. -> "python derslerimiz hakkında" -> ["python","derslerimiz","hakkında"]

ps = PorterStemmer()

def clean_text(text):
    text = text.lower() # tüm metni küçük harfe çevirdi

    text = text.translate(str.maketrans('','', string.punctuation)) # tüm noktalama işaretlerini kaldır..

    words = text.split()

    words = [ word for word in words if not word.isdigit() ]

    words = [ ps.stem(word) for word in words if word not in stopwords.words('english') ]

    return " ".join(words)

clean_text("1. PythON derslerimiz,. hakkında 30")
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

X = df["clean_text"]
y = df["label_num"]

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df["clean_text"])

# [ 
#  [ 0.8, 0.5, 0.3, 0 ,0 ,0],
#  [ 0.4, 1, 0.2, 0 , 0]
# ]

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)
```

```python
#Naive Bayes
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

Modelin Performansını Ölçmek için Metrikler:
- Doğruluk (Accuracy) – Genel olarak tahminlerin doğruluk oranını gösterir.
- Kesinlik (Precision) – Yanlış pozitif tahminleri önlemeye yönelik performansı gösterir.
- Duyarlılık (Recall) – Gerçek pozitifleri doğru şekilde yakalama oranını temsil eder.
- F1-Score – Precision ve Recall’in harmonik ortalamasıdır.
- ROC-AUC – Modelin sınıflandırma yeteneğini gösteren eğri alanı


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='weighted'))
print("Recall:", recall_score(y_test, y_pred, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
```

Modeli Geliştirmek için Adımlar
- Ön işleme ve veri temizliği sürecini iyileştirerek modelin girdi kalitesini artırmak.
- Daha dengeli bir veri seti oluşturarak modelin aşırı öğrenme yapmasını önlemek.
- Ensemble yöntemler ile diğer modellerle birleştirerek doğruluğu artırmak.

KNN Algoritması
KNN (K-Nearest Neighbors), bir noktanın sınıfını belirlemek için en yakın K komşusunun sınıfına bakarak çalışan bir algoritmadır.

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

y_pred_knn = knn_model.predict(X_test)

print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
```

KNN ile performansı artırmak için:
- En iyi K değerini belirlemek (Genelde çapraz doğrulama ile seçilir).
- Özellik ölçeklendirme (Standardization) yaparak mesafe hesaplamalarını daha doğru hale getirmek.
- Farklı mesafe ölçüleri (Euclidean, Manhattan) deneyerek modelin performansını test etmek.
